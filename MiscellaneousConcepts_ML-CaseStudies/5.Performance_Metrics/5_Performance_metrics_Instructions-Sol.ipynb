```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn import metrics
from sklearn.metrics import roc_auc_score
df=pd.read_csv('5_b.csv')
df['y_pred'] = df['proba'].apply(lambda x: 0 if x<0.5 else 1)
y_pred=df['y_pred'].values
y=df['y'].values
print(confusion_matrix(y, y_pred))
print(precision_recall_fscore_support(y, y_pred ))
print(f1_score(y, y_pred, average='micro'))
#fpr, tpr, thresholds = metrics.roc_curve(y, y_pred, pos_label=2)
print(roc_auc_score(y, y_pred))
#print(metrics.auc(fpr, tpr))
```

```python
def confuse_matrix(df):
    tn=0
    tp=0
    fn=0
    fp=0
    y=df['y'].values.tolist()
    y_pred=df['y_pred'].values.tolist()
    for i in range(0,len(df)):
        if(y[i]==y_pred[i]):
            if(y[i]==1):
                tp+=1
            else:
                tn+=1
        elif(y[i]==1 and y_pred[i]==0):
            fn+=1
        else:
            fp+=1
    tpr=tp/(tp+fn)
    fpr=fp/(fp+tn)
    precision=tp/(tp+fp)
    recall=tp/(tp+fn)
    accuracy=(tp+tn)/(tp+tn+fp+fn)
    print("precision:",precision)
    print("recall:",recall)
    F1 = 2 * (precision * recall) / (precision + recall)
    return np.array([tn,fp,fn,tp]).reshape(2,2),F1,tpr,fpr,accuracy
```

```python
df=pd.read_csv('5_a.csv')
threshold_prob=df['proba'].unique().tolist()
threshold_prob=sorted(threshold_prob)
total_tpr=[]
total_fpr=[]
#print(len(sorted(threshold_prob)))
for threshold in threshold_prob:
    df['y_pred'] = df['proba'].apply(lambda x: 0 if x<threshold else 1)
    
    cm,F1,tpr,fpr,accuracy=confuse_matrix(df)
    print("Accuracy=",accuracy)
    total_tpr.append(tpr)
    total_fpr.append(fpr)

#tp=[1,1]tn=[0,0]fn=[1,0]fp=[0,1]
```

```python
total_tpr=sorted(total_tpr)
```

```python
total_fpr=sorted(total_fpr)
```

```python
auc=np.trapz(total_tpr, total_fpr)
auc
```

<font color='red'><b>C.</b></font> Compute the best threshold (similarly to ROC curve computation) of probability which gives lowest values of metric <b>A</b> for the given data <strong>5_c.csv</strong>
<br>

you will be predicting label of a data points like this: $y^{pred}= \text{[0 if y_score < threshold  else 1]}$

$ A = 500 \times \text{number of false negative} + 100 \times \text{numebr of false positive}$

<pre>
   <b>Note 1:</b> in this data you can see number of negative points > number of positive points
   <b>Note 2:</b> use pandas or numpy to read the data from <b>5_c.csv</b>
</pre>

```python
def confuse_matrix(df):
    tn=0
    tp=0
    fn=0
    fp=0
    y=df['y'].values.tolist()
    y_pred=df['y_pred'].values.tolist()
    for i in range(0,len(df)):
        if(y[i]==y_pred[i]):
            if(y[i]==1):
                tp+=1
            else:
                tn+=1
        elif(y[i]==1 and y_pred[i]==0):
            fn+=1
        else:
            fp+=1
    tpr=tp/(tp+fn)
    fpr=fp/(fp+tn)
    precision=tp/(tp+fp)
    recall=tp/(tp+fn)
    accuracy=(tp+tn)/(tp+tn+fp+fn)
    A=(500*fn)+(100*fp)
    print("precision:",precision)
    print("recall:",recall)
    F1 = 2 * (precision * recall) / (precision + recall)
    return np.array([tn,fp,fn,tp]).reshape(2,2),F1,precision,recall,tpr,fpr,accuracy,A
```

```python
df=pd.read_csv('5_c.csv')
df
```

```python
df=pd.read_csv('5_c.csv')
threshold_prob=df['prob'].unique().tolist()
threshold_prob=sorted(threshold_prob)
total_tpr=[]
total_fpr=[]
min_a=1e7
#print(len(sorted(threshold_prob)))
for threshold in threshold_prob:
    df['y_pred'] = df['prob'].apply(lambda x: 0 if x<threshold else 1)
    
    cm,F1,precision,recall,tpr,fpr,accuracy,A=confuse_matrix(df)
    print("Accuracy=",accuracy)
    print("A Value=",A)
    if(min_a>A):
        min_a=A
        optimal_threshold=threshold
        optimal_precision=precision
        optimal_recall=recall
        optimal_F1=F1
        optimal_accuracy=accuracy
    
    total_tpr.append(tpr)
    total_fpr.append(fpr)

#tp=[1,1]tn=[0,0]fn=[1,0]fp=[0,1]
```

```python
print("Lowest A value: ",min_a)
print("Optimal_Threshold:",optimal_threshold)
print("Optimal_Precision:",optimal_precision)
print("Optimal_Recall:",optimal_recall)
print("Optimal_F1:",optimal_F1)
print("Optimal_Accuracy:",optimal_accuracy)
```

```python
total_tpr=sorted(total_tpr)
total_fpr=sorted(total_fpr)
auc=np.trapz(total_tpr, total_fpr)
auc
```

```python
import logging
def mean_absolute_percentage_error(y_true, y_pred): 
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    except error as e:
        logging.error(e)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
df=pd.read_csv('5_d.csv')
errors = mean_squared_error(df['y'], df['pred'])
#errors_perc = mean_absolute_percentage_error(df['y'], df['pred'])
r2 = r2_score(df['y'], df['pred'])
print(errors)
print('r2 score for the model is', r2)
```
